```python
import pandas as pd
import nltk
import gensim 
```

```python
train_raw = pd.read_csv('train_raw.csv', index_col=0)
train_raw.head()
```

```python
words = [gensim.utils.simple_preprocess(w) for w in train_raw.review_body.dropna().tolist()]
```

```python
words
```

```python
# build vocabulary and train model
model = gensim.models.Word2Vec(
    words,
    size=150,
    window=10,
    min_count=2,
    workers=10,
    iter=10)
```

```python
model.wv.most_similar(positive='monitor', topn=6)
```
