```python
def isNaN(string):
    return string != string

def clean(review):
    if (isNaN(review)):
        return ''
    # function to test if something is a noun
    is_noun = lambda pos: pos[:2] == 'NN'
    tokenized = nltk.word_tokenize(review)
    ps = PorterStemmer()
    nouns = [ps.stem(word) for (word, pos) in nltk.pos_tag(tokenized) if is_noun(pos)]
    return ' '.join(nouns)
```

```python
def create_agg(df, col):
    criteria = {'review_id': 'count', 
            'star_rating': ['mean', 'std'], 
            'total_votes': ['mean', 'std'],
            'helpful_votes': ['mean', 'std']}
    return df.groupby(df[col]).agg(criteria).fillna(0)
```

```python
def concat_corpus(source_df, product_parent):
    return ' '.join(source_df.query('product_parent == {}'.format(product_parent)).clean_review)
```

```python
def pos_and_neg(source_df, customer_id):
    pos_words = ' '.join(source_df.query('customer_id == {} & positive == 1'.format(customer_id)).clean_review)
    neg_words = ' '.join(source_df.query('customer_id == {} & positive == 0'.format(customer_id)).clean_review)
    return [pos_words, neg_words]
```

```python
def pos(source_df, customer_id):
    return ' '.join(source_df.query('customer_id == {} & positive == 1'.format(customer_id)).clean_review)
```

```python
def neg(source_df, customer_id):
    return ' '.join(source_df.query('customer_id == {} & positive == 0'.format(customer_id)).clean_review)
```

```python
import heapq
def get_top_3(corpus):
    res = ['', '', '']
    words = str(corpus).split()
    words = {w : words.count(w) for w in words}
    
    heap_array = heapq.nlargest(3, words, key=words.get)
    i = 0
    for w in heap_array:
        res[i] = w
        i += 1
        
    return res
```

```python
train_pos = train_cust_df.pos_corpus.apply(get_top_3)
train_neg = train_cust_df.neg_corpus.apply(get_top_3)
train_cust_df.head()
```

```python
def jaccard_similarity(query, document):
    query = query.split()
    document = document.split()
    intersection = set(query).intersection(set(document))
    union = set(query).union(set(document))
    if (len(union) == 0):
        return 0
    return len(intersection)/len(union)
```

```python
train_df['pos_sim'] = train_df.apply((lambda x: jaccard_similarity(x.pos_corpus, x.prod_corpus)), axis=1)
train_df['neg_sim'] = train_df.apply((lambda x: jaccard_similarity(x.neg_corpus, x.prod_corpus)), axis=1)
train_df.head()
```
