```python
import pandas as pd
import numpy as np

df = pd.read_csv('cluster_raw.csv', index_col=0)
df.head()
```

```python
import skfuzzy as fuzz

def generate_models(df):
    models = []
    for num_clusters in range(2, 8):
        print('\n---', num_clusters, '---')
        cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(df, num_clusters, 2, 0.005, 1000)
        models.append((cntr, u, u0, d, jm, p, fpc))
        print('FPC:', fpc)
        cluster_membership = np.argmax(u, axis=0)
        for i  in range(num_clusters):
            cluster = cluster_membership[cluster_membership == i]
            print('Cluster {}: {}%'.format(i, (len(cluster) / len(cluster_membership)) * 100))
    return models
```

```python
model_sets = []

criteria = {
    'review_id': 'count', 
    'star_rating': ['mean', 'std'], 
    'total_votes': ['mean', 'std'],
    'helpful_votes': ['mean', 'std']
}
swapped1 = np.swapaxes(df.groupby('customer_id').agg(criteria).fillna(0), 0,1)
model_sets.append(generate_models(swapped1))
```
